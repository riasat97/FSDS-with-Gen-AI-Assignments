## 1. Email Validator (String Filtering + Loops)
- Real-world use: Used in signup forms or data cleaning tasks.
Task:
- Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.
- Example:
- Input: hello@inceptionbd.com
- Output: Valid Email

```python
email= input("Enter your E-mail Address")
check_at= email.find("@")
email_endsWith_com= email.endswith(".com")
email_endsWith_org= email.endswith(".org")
if check_at!=-1 and (email_endsWith_com or email_endsWith_org):
    print("Valid Email")
else:    
    print("Error! Invalid Email.")
```

## 2. Count Word Frequency in Customer Reviews
Real-world use: Analyzing reviews or comments in text analytics.
Task:
- Given a review text, count how many times each word appears (ignore case).
- Example:
- Input: "This product is good and this service is excellent"
- Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
text= input("Review text").lower()
text_list= text.split()
unique_words=[]
word_counts=[]
for word in text_list:
    if word not in unique_words:
        unique_words.append(word)
        word_counts.append(1)
    else:
        index=unique_words.index(word)
        word_counts[index]+=1
# print(unique_words)
# print(word_counts)
string_pairs=[]
for index,word in enumerate(unique_words):
    string=f"'{word}': {word_counts[index]}"
    string_pairs.append(string)
#print(string_pairs)
inner_content=", "
print("{"+ inner_content.join(string_pairs)+"}")


```

## 3. Password Strength Checker
Real-world use: Used in user authentication systems.
- Task:
- Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.
- Example:
- Input: "Hello@123"
- Output: Strong Password

```python

strong=False
while not strong:
 password=input('Enter password: ')
 uppercase=0 
 lowercase=0
 digit=0
 special_char=0

 for ch in password:
    if ch.isupper():
        uppercase+=1
    elif ch.islower():
        lowercase+=1
    elif ch.isdigit():
        digit+=1
    elif not ch.isalnum():
        special_char+=1
#print(f"{uppercase}, {lowercase}, {digit}, {special_char}")    
 if uppercase and lowercase and digit and special_char:
    strong=True
 else:
   print('Weak password. Must contain uppercase, lowercase, a digit, and a special character.')
   print('Please try again.')
else: 
   print('Strong password accepted!')
```

## 4. Username Generator
Real-world use: Automatically create usernames for users.
- Task:
- Take user‚Äôs full name and generate a short username.
- Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
- Example:
- Input: "Mamun Malitha"
- Output: mamitha57

```python
name= input("Full name").lower()
name_split= name.split()
first_name= name_split[0]
last_name= name_split[len(name_split)-1]
number_total=0
for ch in name:
    number_total+=ord(ch)
#print(number_total)
final_number= number_total%100    
print(first_name[0:3]+last_name[-3:]+str(final_number))
#username= name[0:3]
```

## 6. Remove Duplicates from Contact List
Real-world use: Data cleaning in CRM or user databases.
- Task:
- Given a list of names (comma separated), remove duplicates.
- Example:
- Input: "Rahim, Karim, Rahim, Sultana"
- Output: "Rahim, Karim, Sultana"

```python
contact_list= input("Enter a list of names (comma separated)")
contact_list_split= contact_list.split(",")
unique_contacts=[]
for contact in contact_list_split:
      cleaned_name = contact.strip()
      if cleaned_name not in unique_contacts:
          unique_contacts.append(cleaned_name)
final_list=", "           
print(final_list.join(unique_contacts))
```

## 7. Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
- Task:
- If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
- Example:
- Input: "This is a scam offer"
- Output: Warning: Suspicious content detected

```python
text= input("Enter message: ")
text_split= text.split()
banned_words= ['spam','scam','fake']
suspicious=False
for word in text_split:
    if word in banned_words:
        suspicious= True
print("Warning: Suspicious content detected" if suspicious else "Accepted")        
```

## 10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
- Task:
- Reverse each word in a sentence while keeping word order the same.
- Example:
- Input: "AI is fun"
- Output: "IA si nuf"


```python
message= input("Enter message: ")
text_split= message.split()
reversed_words = [ word[::-1] for word in text_split]
encripted_text=" ".join(reversed_words)
print(encripted_text)
```
